```python
from src.tokenizers.BPE import BytePairEncoding
from src.utils import paths
```

```python
path_tokenizer_en = paths.EXPERIMENTS_DIR / "multi30k/tokenizer_en.json"
path_tokenizer_de = paths.EXPERIMENTS_DIR / "multi30k/tokenizer_de.json"

tokenizer_en = BytePairEncoding.from_file(path_tokenizer_en)
tokenizer_de = BytePairEncoding.from_file(path_tokenizer_de)
```

```python
print(tokenizer_en.tokenize("english"))
print(tokenizer_de.tokenize("deutch"))

print(tokenizer_en.decode(tokenizer_en.tokenize("english")))
print(tokenizer_de.decode(tokenizer_de.tokenize("deutch")))
```

```python
print(tokenizer_en.tokenize("english", max_length=3))
print(tokenizer_de.tokenize("deutch", max_length=3))

print(tokenizer_en.decode(tokenizer_en.tokenize("english", max_length=3), add_special=True))
print(tokenizer_de.decode(tokenizer_de.tokenize("deutch", max_length=3), add_special=True))
```

```python
print(tokenizer_en.tokenize("english", max_length=10, pad=True))
print(tokenizer_de.tokenize("deutch", max_length=10, pad=True))

print(tokenizer_en.decode(tokenizer_en.tokenize("english", max_length=10, pad=True), add_special=True))
print(tokenizer_de.decode(tokenizer_de.tokenize("deutch", max_length=10, pad=True), add_special=True))
```

```python
print(tokenizer_en.tokenize("A slightly longer sentence", max_length=20, pad=True))
print(tokenizer_de.tokenize("A slightly longer sentence", max_length=20, pad=True))

print(tokenizer_en.decode(tokenizer_en.tokenize("A slightly longer sentence", max_length=20, pad=True), add_special=True))
print(tokenizer_de.decode(tokenizer_de.tokenize("A slightly longer sentence", max_length=20, pad=True), add_special=True))
```

```python
print(tokenizer_en.tokenize("  !  ", max_length=20, pad=True))
print(tokenizer_de.tokenize("  !  ", max_length=20, pad=True))

print(tokenizer_en.decode(tokenizer_en.tokenize("  !  ", max_length=20, pad=True), add_special=True))
print(tokenizer_de.decode(tokenizer_de.tokenize("  !  ", max_length=20, pad=True), add_special=True))
```
